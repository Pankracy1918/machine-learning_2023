```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('./dataset/WHR_2023.csv')
```

```python
data.head()

hs_frmfc = data[['happiness_score', 'freedom_to_make_life_choices']]
```

```python
sns.lmplot(x="happiness_score", y="freedom_to_make_life_choices", data=hs_frmfc, order=2, ci=None)
```

```python
hs_frmfc[['freedom_to_make_life_choices', 'happiness_score']].isna().sum()
```

```python
X = hs_frmfc['freedom_to_make_life_choices'].values.reshape(-1, 1)
y = hs_frmfc['happiness_score'].values.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
regr = LinearRegression()

regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)
plt.scatter(X_test, y_test, color='b')
plt.plot(X_test, y_pred, color = 'k')

plt.show()
```

```python
#EVAL
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_squared_error(y_true=y_test, y_pred=y_pred)

mse = mean_squared_error(y_true=y_test, y_pred=y_pred)

rmse = mean_squared_error(y_true=y_test, y_pred=y_pred, squared=False)

print(mae)
print(mse)
print(rmse)
```
